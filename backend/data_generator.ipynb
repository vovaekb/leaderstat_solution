```python
import pandas as pd
import random
from datetime import datetime
import time
import json
import numpy as np
```

```python
def randomDate(start, end):
    frmt = '%d-%m-%Y %H:%M:%S'

    stime = time.mktime(time.strptime(start, frmt))
    etime = time.mktime(time.strptime(end, frmt))

    ptime = stime + random.random() * (etime - stime)
    dt = datetime.fromtimestamp(time.mktime(time.localtime(ptime)))
    return dt

```

```python
Location = ['Фили', 'Карамышевский', 'Верхние Мнёвники', 'Сидней-Сити', 'Хруничева', "Сердце столицы", "Мелькомбинат", "Западный порт","Кутузовский", "Москва-Сити", "Мост Багратион", "ТЭЦ", "Киевский вокзал", "Автозаводский мост", "ЗИЛ","Нагатинская набережная","ЗИЛ-Юг","Южный речной вокзал", "Кленовый бульвар", "Негатинский затон", "Коломенская набережная", "Печатники", "Южный грузовой порт"]
max_count = [random.randrange(500, 2000, 100) for i in range(len(Location))]

j = 0
dict_max_loc ={}
for j in range(len(Location)):
    dict_max_loc[Location[j]] = max_count[j]
    j += 1
    
```

```python
dict_max_loc
```

```python
columns = ['DataTime', 'Location'] # Датафрейм для причалов
data = {'DataTime': [randomDate("21-12-2021 13:30:00", "23-12-2021 07:50:00") for i in range(2000)],       
       'Location':[random.choice(Location) for i in range(2000)]
       }
```

```python
data = pd.DataFrame(data=data, columns=columns)
data.head()
```

```python
data['Max_count'] = data['Location'].map(dict_max_loc)
```

```python
data['count_pass'] = [random.uniform(0.2*data['Max_count'][i], data['Max_count'][i]) for i in range(2000)]
```

```python
data['count_pass'] = data['count_pass'].apply(np.int64)
```
